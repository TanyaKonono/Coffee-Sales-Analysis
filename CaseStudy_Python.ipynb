```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the CSV file into a dataframe called df
df = pd.read_csv("/Volumes/workspace/bright_coffee_sales/casestudyfiles/Bright Coffee Shop Analysis - Transactions.csv")
```

```python
# Step 2: Look at the first 5 rows so you can see the data
print(df.head())
```

```python
# Step 3: Convert the date column into a proper date format
df["transaction_date"] = pd.to_datetime(df["transaction_date"], errors="coerce")

#erros="coerce"If a value can be converted to a number, convert it. If a value cannot be converted (it raises an error),
# replace it with NaN (Not a Number).

# Step 4: Convert the time column into a proper time format
df["transaction_time"] = pd.to_datetime(df["transaction_time"], format="%H:%M:%S", errors="coerce")

# Step 5: Create a new column called sales_amount
# This is quantity × unit price
df["sales_amount"] = df["transaction_qty"] * df["unit_price"]

# Step 6: Create a new column for the hour
# This helps us analyse which hours are busiest
df["hour"] = df["transaction_time"].dt.hour

# Step 7: Create a new column for the month
# This helps us analyse monthly sales trends
df["month"] = df["transaction_date"].dt.to_period("M").astype(str)


```

```python
# -------------------------------
# VISUAL 1: Monthly Sales Trend
# -------------------------------

# Group the data by month and sum the sales
monthly_sales = df.groupby("month")["sales_amount"].sum()

# Create the graph
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker="o")

# Add title and labels
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")

# Rotate x-axis labels so they are easy to read
plt.xticks(rotation=45)

# Make layout neat
plt.tight_layout()

# Show the graph
plt.show()
```

```python
# -----------------------------------
# VISUAL 2: Sales by Product Category
# -----------------------------------

# Group the data by product category and sum the sales
category_sales = df.groupby("product_category")["sales_amount"].sum().sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 5))
plt.bar(category_sales.index, category_sales.values)

# Add title and labels
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Sales Amount")

# Rotate labels for readability
plt.xticks(rotation=45)

# Make layout neat
plt.tight_layout()

# Show the graph
plt.show()


```

```python
# --------------------------------
# VISUAL 3: Sales by Store Location
# --------------------------------

# Group the data by store location and sum the sales
location_sales = df.groupby("store_location")["sales_amount"].sum().sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(location_sales.index, location_sales.values)

# Add title and labels
plt.title("Sales by Store Location")
plt.xlabel("Store Location")
plt.ylabel("Sales Amount")

# Rotate labels slightly
plt.xticks(rotation=20)

# Make layout neat
plt.tight_layout()

# Show the graph
plt.show()
```

```python
# -----------------------------
# VISUAL 4: Transactions by Hour
# -----------------------------

# Count how many transactions happened in each hour
hourly_transactions = df.groupby("hour")["transaction_id"].count()

# Create the line chart
plt.figure(figsize=(10, 5))
plt.plot(hourly_transactions.index, hourly_transactions.values, marker="o")

# Add title and labels
plt.title("Transactions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")

# Show every hour on the x-axis
plt.xticks(hourly_transactions.index)

# Make layout neat
plt.tight_layout()

# Show the graph
plt.show()
```

```python
# -------------------------------------
# VISUAL 5: Top 10 Product Types by Sales
# -------------------------------------

# Group by product type, sum sales, sort highest to lowest, then keep top 10
top_product_types = (
    df.groupby("product_type")["sales_amount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Create the bar chart
plt.figure(figsize=(11, 6))
plt.bar(top_product_types.index, top_product_types.values)

# Add title and labels
plt.title("Top 10 Product Types by Sales")
plt.xlabel("Product Type")
plt.ylabel("Sales Amount")

# Rotate labels so they fit
plt.xticks(rotation=45, ha="right")

# Make layout neat
plt.tight_layout()

# Show the graph
plt.show()
```
